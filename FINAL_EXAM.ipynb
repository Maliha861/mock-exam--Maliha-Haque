```python
# Section 1: Iterative Prompt Engineering (30 marks)

## 1.1 Initial Prompt & Pseudocode (8 marks)

#PROMPT 1
I need to create a command-line interface task manager in Python for my introduction to programming course.
The programme should allow users to add, list, and remove tasks that are stored in memory.
Kindly provide pseudocode that only uses basic Python concepts, such as basic lists and dictionaries for data storage.
It should have simple functions with parameters suitable for our introductory course with no advanced features.
Please follow a structured planning approach showing clear steps.

Six-step process to design a programme:

Step 1: Understand the problem
Design an interactive command-line task manager that allows users to manage simple text-based tasks through a numbered menu.
The program lets users:
-Add a new task (with a title)
-List all current tasks
-Remove a task by its ID

Step 2:Identify the input and output
Inputs: menu choice (string), task title (string), task ID for removal (string → parsed to int)
Outputs: printed confirmations and a table of tasks (ID | Title)

Step 3:Work the problem by hand:
Start with no tasks: `tasks = []`
User adds two tasks — “Buy milk” and “Buy groceries”
System stores them as:
`[{"id":1,"title":"Buy milk"}, {"id":2,"title":"Buy groceries"}]`
User lists tasks and sees both tasks on screen
User removes task with ID 1 (“Buy milk”)
System updates list to keep only task 2 (“Buy groceries”)
User quits the program

Step 4:Write the pseudocode:
FUNCTION show_menu():
    DISPLAY "1) Add Task"
    DISPLAY "2) List Tasks"
    DISPLAY "3) Remove Task"
    DISPLAY "q) Quit"

FUNCTION add_task(tasks, next_id, title):
    IF title is empty:
        DISPLAY "Title cannot be empty."
        RETURN next_id
    CREATE new_task = {"id": next_id, "title": title, "done": False}
    ADD new_task to tasks
    DISPLAY "Task added with ID", next_id
    RETURN next_id + 1

FUNCTION list_tasks(tasks):
    IF length of tasks is 0:
        DISPLAY "No tasks yet."
    ELSE:
        DISPLAY "ID | Title"
        FOR each task in tasks:
            DISPLAY task["id"], "|", task["title"]

FUNCTION remove_task(tasks, id_text):
    TRY to convert id_text to integer → target_id
    IF conversion fails:
        DISPLAY "Invalid ID. Please enter a number."
        RETURN
    SET found = False
    FOR each task in tasks:
        IF task["id"] equals target_id:
            REMOVE task from tasks
            found = True
            DISPLAY "Task removed."
            BREAK
    IF found is False:
        DISPLAY "No task found with that ID."

MAIN PROGRAMME:
    CREATE empty list tasks
    SET next_id = 1
    SET running = True

    WHILE running is True:
        CALL show_menu()
        GET choice from user

        IF choice equals "1":
            GET title from user
            SET next_id = add_task(tasks, next_id, title)

        ELSE IF choice equals "2":
            CALL list_tasks(tasks)

        ELSE IF choice equals "3":
            GET id_text from user
            CALL remove_task(tasks, id_text)

        ELSE IF choice equals "q":
            DISPLAY "Goodbye!"
            SET running = False

        ELSE:
            DISPLAY "Invalid option. Please try again."

#Course Reference:
From the six-step process to design a programme taught in Week 2, Module 3, Computational Thinking and Algorithm Design.
I used the four key stages — planning, implementation, testing, and evaluation.
I used it as an outline for my work to organise the pseudocode into logical steps for creating the Task Manager program.
```
